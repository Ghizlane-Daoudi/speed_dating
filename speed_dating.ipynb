### Create a dataset with only characteristics of participants

```python
# Characteristics of individuals not dependending on the date experience
participant_profile_columns = ['age','gender','race', 'field_cd','career_c','from','zipcode',
                               'goal','date','go_out', 'imprace', 'imprelig',
                               'exphappy','expnum',
                               'sports','tv','sports','exercise','dining', 'museums','art','hiking','gaming',
                               'clubbing','reading','tv','theater','movies','concerts','music','shopping','yoga']

particip_profile = pd.DataFrame()

for column in participant_profile_columns:
    if type(dataset[column][0]) in ('float64','int64') :
        particip_profile[column] = dataset.groupby('iid')[column].median()
    else:
        particip_profile[column] = dataset.groupby('iid')[column].max()
    
particip_profile.reset_index(inplace=True)
particip_profile.head()
```

### Merged dataset with all columns for both participants

```python
# Creating a merged dataset to have a complete overview on both participants for each date
merged_dataset = dataset.merge(dataset, how='inner', left_on=['iid','pid'], right_on=['pid','iid'], suffixes=('', '_partner')) 
```

```python
merged_dataset.head()
```

```python
# Extract synthetic criteria to compare against matches between participants

merged_dataset['Age difference']=abs(merged_dataset['age']-merged_dataset['age_partner'])
merged_dataset['Date frequency']=abs(merged_dataset['date']-merged_dataset['date_partner'])
merged_dataset['Go out frequency']=abs(merged_dataset['go_out']-merged_dataset['go_out_partner'])
merged_dataset['Studies field']=[1 if merged_dataset['field_cd'][i]==merged_dataset['field_cd_partner'][i] 
                                else 0 for i in range(merged_dataset.shape[0])]
merged_dataset['Pretended career']=[1 if merged_dataset['career_c'][i]==merged_dataset['career_c_partner'][i] 
                                else 0 for i in range(merged_dataset.shape[0])]
merged_dataset['Origin area']=[1 if merged_dataset['from'][i]==merged_dataset['from_partner'][i] 
                                else 0 for i in range(merged_dataset.shape[0])]
merged_dataset['Zip Code']=[1 if merged_dataset['zipcode'][i]==merged_dataset['zipcode_partner'][i] 
                                else 0 for i in range(merged_dataset.shape[0])]
merged_dataset['Goal']=[1 if merged_dataset['goal'][i]==merged_dataset['goal_partner'][i] 
                                else 0 for i in range(merged_dataset.shape[0])]
```

### Studies and career analysis

```python
dict_studies = {0:'Different studies field', 1:'Same studies field'}
dict_career = {0:'Different pretended career', 1:'Same pretended career'}
dict_origin = {0:'From different areas', 1:'From the same area'}
dict_samegoal = {0:'Different goal', 1:'Same goal'}
dict_frequency = {0:'Different frequency', 1:'Same frequency'}
```

```python
# Creation of two graphics to see if having the same career, studies or origin impacts the matching between participants
plt.figure(figsize=(20,5))

plt.subplot(131)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Studies field'].map(dict_studies))
plt.title('Matches according to the studies field')

plt.subplot(132)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Pretended career'].map(dict_career))
plt.title('Matches according to the pretended career')

plt.subplot(133)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Origin area'].map(dict_origin))
plt.title('Matches according to the origin area')

plt.subplots_adjust(wspace=0.5)
```

### Participants goal and social activities analysis

```python
# Creation of two graphics to see if having the same goal, goingout or dating frequency impacts the matching between participants
plt.figure(figsize=(20,5))

plt.subplot(131)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Go out frequency'].map(dict_frequency))
plt.title('Matches according to the going out frequency')

plt.subplot(132)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Date frequency'].map(dict_frequency))
plt.title('Matches according to dating frquency')

plt.subplot(133)
sns.countplot(merged_dataset['match'].map(dict_match),hue=merged_dataset['Goal'].map(dict_samegoal))
plt.title('Matches according to the dating goal')

plt.subplots_adjust(wspace=0.5)
```
